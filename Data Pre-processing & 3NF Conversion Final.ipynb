**Create Post-season and Regular-season binary label**

Upon examining the data, we observe that the 'W' and 'L' in the table do not change after the regular season ends. However, because each team finishes their regular seasons on different dates and the total 'W' + 'L' is not always 82 (i.e., during the Covid period), it is not apparent when the post-season begins. Therefore, we decide to use the logic of the first day that the 'W' + 'L' no longer increases in a season as a mark for when the post-season begins and create a binary label.

```python
Ranking_cleaned = ranking_new.copy()
Ranking_cleaned['Total_Games'] = Ranking_cleaned['W'] + Ranking_cleaned['L']
max_games_played = Ranking_cleaned.groupby(['TEAM_ID', 'SEASON_ID'])['Total_Games'].max().reset_index()
Ranking_cleaned = Ranking_cleaned.merge(max_games_played, on = ['TEAM_ID', 'SEASON_ID'])
Ranking_cleaned = Ranking_cleaned.rename(columns = {'Total_Games_x': 'Total_by_Date', 'Total_Games_y': 'Total_Games_Max_Season'})
Ranking_cleaned
```

```python
# find the end of season date for each team in each season
end_of_season = Ranking_cleaned[Ranking_cleaned['W'] + Ranking_cleaned['L'] == Ranking_cleaned['Total_Games_Max_Season']][['TEAM_ID', 'SEASON_ID','STANDINGSDATE']].groupby(['TEAM_ID', 'SEASON_ID'])['STANDINGSDATE'].min().reset_index()
end_of_season
```

```python
# sanity check
end_of_season[end_of_season['TEAM_ID'] == 1610612737]
```

```python
temp = Ranking_cleaned.merge(end_of_season, on = ['TEAM_ID', 'SEASON_ID'])
temp['is_out_of_regular_season'] = temp['STANDINGSDATE_x'] > temp['STANDINGSDATE_y']
temp.shape
```

```python
RANKING_FINAL = temp[['TEAM_ID', 'LEAGUE_ID', 'SEASON_ID', 'STANDINGSDATE_x', 'CONFERENCE', 'TEAM', 'G', 'W', 'L', 'W_PCT','HOME_RECORD', 'ROAD_RECORD', 'RETURNTOPLAY', 'is_out_of_regular_season']]
RANKING_FINAL = RANKING_FINAL.rename(columns = {'STANDINGSDATE_x': 'STANDINGSDATE'})

# finally the LEAGUE_ID AND RETURNTOPLAY columns are not useful, therefore remove
RANKING_FINAL = RANKING_FINAL.drop(columns = ['LEAGUE_ID', 'RETURNTOPLAY'])
RANKING_FINAL.head()
```

## Part 2: 3NF & Entity Resolution

With players, teams, ranking, games, and games_details cleaned. We now need to ensure all the relations are in 3NF.

### 2.1 Players

```python
player_new.columns
```

Currently the relation is: Player(<u>TEAM_ID</u>, <u>PLAYER_ID</u>, <u>SEASON</u>, PLAYER_NAME) <br>
With FD: PLAYER_ID -> PLAYER_NAME

To conform to 3NF, we need to decompose it into: <br>
Roster(<u>TEAM_ID</u>, <u>PLAYER_ID</u>, <u>SEASON</u>) <br>
Player(<u>PLAYER_ID</u>, PLAYER_NAME)

```python
roster_3nf = player_new[['TEAM_ID', 'PLAYER_ID', 'SEASON']]
roster_3nf = roster_3nf.drop_duplicates()

player_3nf = player_new[['PLAYER_ID', 'PLAYER_NAME']]
player_3nf = player_3nf.drop_duplicates()
player_3nf.head()
```

```python
teams.head()
```

### 2.3 Games

```python
games.columns
```

Currently the relation is: Games(<u>GAME_ID</u>, GAME_DATE_EST, SEASON, TEAM_ID_home, PTS_home, FG_PCT_home, FT_PCT_home, FG3_PCT_home, AST_home, REB_home, TEAM_ID_away, PTS_away, FG_PCT_away, FT_PCT_away, FG3_PCT_away, AST_away, REB_away) <br>

With FD: GAME_ID -> GAME_DATE_EST, SEASON, TEAM_ID_home, PTS_home, FG_PCT_home, FT_PCT_home, FG3_PCT_home, AST_home, REB_home, TEAM_ID_away, PTS_away, FG_PCT_away, FT_PCT_away, FG3_PCT_away, AST_away, REB_awa

The resulting relation already conforms to 3NF, no transitive / partial dependency. Out of consideration that sometimes we don't need the exact points scored by teams but just the season and date the game is played in, the relation is further decomposed into: <br>
Games(<u>GAME_ID</u>, GAME_DATE_EST) <br>
Game_Stats_Team(<u>GAME_ID</u>, TEAM_ID_home, PTS_home, FG_PCT_home, FT_PCT_home, FG3_PCT_home, AST_home, REB_home, TEAM_ID_away, PTS_away, FG_PCT_away, FT_PCT_away, FG3_PCT_away, AST_away, REB_away)

```python
games_3nf = games[['GAME_ID', 'GAME_DATE_EST', 'SEASON']].drop_duplicates()
games_3nf.shape
```

```python
game_stats_team_3nf = games[['GAME_ID', 'HOME_TEAM_ID',
       'VISITOR_TEAM_ID', 'SEASON', 'TEAM_ID_home', 'PTS_home', 'FG_PCT_home',
       'FT_PCT_home', 'FG3_PCT_home', 'AST_home', 'REB_home', 'TEAM_ID_away',
       'PTS_away', 'FG_PCT_away', 'FT_PCT_away', 'FG3_PCT_away', 'AST_away',
       'REB_away']].drop_duplicates()
game_stats_team_3nf.shape
```

### 2.4 games_details

```python
games_details_new.columns
```

-> In the current relation
- TEAM_ABBREVIATION, TEAM_CITY can be infered from TEAM_ID using the teams table
- PLAYER_NAME can be infered from PLAYER_ID using the players table
- FG_PCT can be calculated from FGM and FGA
- FG3_PCT can be calculated from FG3M and FG3A
- FT_PCT can be calculated from FTM and FTA
- PTS can also be calculated from FG3M, FGM, and FTM

Therefore, drop all of these unnecessary columns:

```python
games_details_3nf = games_details_new.loc[:,['GAME_ID','TEAM_ID','PLAYER_ID','MIN','FGM','FGA','FG3M','FG3A','FTM','FTA',
                         'OREB','DREB','AST','STL','BLK','TO','PF','PLUS_MINUS']]
games_details_3nf.columns
```

```python
# adding a home_or_away column

temp1 = game_stats_team_3nf[['GAME_ID', 'TEAM_ID_home']]
temp1.columns = ['GAME_ID', 'TEAM_ID']
temp1['home_or_away'] = 'home'
temp1.head()

temp2 = game_stats_team_3nf[['GAME_ID', 'TEAM_ID_away']]
temp2.columns = ['GAME_ID', 'TEAM_ID']
temp2['home_or_away'] = 'away'
temp2.head()

temp_merge = pd.concat([temp1, temp2], axis = 0)
temp_merge = temp_merge.drop_duplicates()
temp_merge.head()

Game_Stats_Player = games_details_3nf.merge(temp_merge, on = ['GAME_ID', 'TEAM_ID'])
Game_Stats_Player.columns
Game_Stats_Player = Game_Stats_Player[['GAME_ID', 'TEAM_ID', 'PLAYER_ID', 'MIN', 'FGM', 'FGA',
       'FG3M', 'FG3A', 'FTM', 'FTA', 'OREB', 'DREB', 'AST', 'STL', 'BLK', 'TO',
       'PF', 'PLUS_MINUS', 'home_or_away']]
```

```python
Game_Stats_Player_3nf = Game_Stats_Player.drop_duplicates()
Game_Stats_Player_3nf.head()
```

Relation: Game_Stats_Player(<u>GAME_ID</u>, <u>TEAM_ID</u>, <u>PLAYER_ID</u>, MIN, FGM, FGA, FG3M, FG3A, FTM, FTA, OREB, DREB, AST, STL, BLK, TO, PF, PLUS_MINUS, home_or_away)

FD: GAME_ID, TEAM_ID, PLAYER_ID -> MIN, FGM, FGA, FG3M, FG3A, FTM, FTA, OREB, DREB, AST, STL, BLK, TO, PF, PLUS_MINUS, home_or_away

There is no transitive or partial dependency, therefore it is in 3NF

### 2.5 Ranking

```python
# drop unnecessary columns
RANKING_FINAL.drop(columns = ['SEASON_ID'], inplace = True)
RANKING_FINAL.columns 
```

```python
RANKING_FINAL.head()
```

Relation: Ranking(<u>TEAM_ID</u>, <u>STANDINGSDATE</u>, CONFERENCE, TEAM, G, W, L, W_PCT, HOME_RECORD, ROAD_RECORD, is_out_of_regular_season) <br>

FDs:<br>
W, L -> W_PCT<br>
W, L -> G <br>
TEAM_ID -> TEAM<br>
TEAM_ID, STANDINGSDATE -> CONFERENCE, TEAM, G, W, L, W_PCT, HOME_RECORD, ROAD_RECORD, is_out_of_regular_season

Since W_PCT and G can be calculated using exisitng data, they will be excluded; Team can also be infered from the current team relation (2.2) so also excluded, resulting in: <br>
Ranking(<u>TEAM_ID</u>, <u>STANDINGSDATE</u>, CONFERENCE, W, L, HOME_RECORD, ROAD_RECORD, is_out_of_regular_season) <br>
with FD: TEAM_ID, STANDINGSDATE -> CONFERENCE, W, L, HOME_RECORD, ROAD_RECORD, is_out_of_regular_season

No transitive or partial dependency, therefore in 3NF

```python
ranking_3nf = RANKING_FINAL.loc[:,['TEAM_ID', 'STANDINGSDATE','CONFERENCE','W','L','HOME_RECORD','ROAD_RECORD']]
ranking_3nf = ranking_3nf.drop_duplicates()
ranking_3nf.shape
```
